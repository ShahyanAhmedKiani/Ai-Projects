```python
from ultralytics import YOLO
import cv2
import os
```

```python

# -----------------------------
# 1. Load the YOLO Model
# -----------------------------
model = YOLO("best.pt")      # or "last.pt"
```

```python



# -----------------------------
# 2. Test on a Single Image
# -----------------------------
def test_single_image(image_path):
    print("\n🔍 Testing Single Image...\n")

    results = model(image_path, save=True, show=True)  
    # save=True → saves output to 'runs/detect/predict'
    # show=True → displays image with bounding boxes

    for r in results:
        print("\n📌 Prediction Details:")
        for box in r.boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            xyxy = box.xyxy[0].tolist()

            print(f"Class: {model.names[cls]}")
            print(f"Confidence: {conf:.4f}")
            print(f"Bounding Box (x1,y1,x2,y2): {xyxy}")

    print("\n✅ Saved output image to runs/detect/predict/")
```

```python






# -----------------------------
# 3. Test All Images in a Folder
# -----------------------------
def test_folder(folder_path):
    print("\n📁 Testing All Images in Folder...\n")

    # List all image files
    images = [img for img in os.listdir(folder_path)
              if img.lower().endswith((".jpg", ".png", ".jpeg"))]

    for img_name in images:
        img_path = os.path.join(folder_path, img_name)
        print(f"\n🔍 Testing: {img_name}")

        results = model(img_path, save=True)

        # Print results
        for r in results:
            for box in r.boxes:
                cls = int(box.cls[0])
                conf = float(box.conf[0])
                print(f" → Class: {model.names[cls]}, Confidence: {conf:.3f}")

    print("\n✅ All processed images saved in runs/detect/predict/")


# -----------------------------
# 4. Run Testing
# -----------------------------

# Test on a single image
test_single_image("1 no.jpeg")   # <--- change to your image

# Test on a folder
# test_folder("test_images")              # <--- change to your folder name
```
